```python
import numpy as np
print(np.__version__)
```

```python
data=np.genfromtxt("heart.csv",delimiter=",",skip_header=True)
age=data[:,0]

#Basic stats

print(f"age: {age[:5]}") #test code
sex=data[:,1]
cp=data[:,2]
trestbps=data[:,3]
chol=data[:,4]
fbs=data[:,5]
restecg=data[:,6]
thalach=data[:,7]
print(f"thalach: {thalach[:5]}")
exang=data[:,8]
oldpeak=data[:,9]
slope=data[:,10]
ca=data[:,11]
thal=data[:,12]
print(f"thal;: {thal[:5]}")
target=data[:,13]
print(f"target: {target[:5]}")

#Basic desciptive stats function 
def descipt_stat(stat):
    
    print(f"Mean: {round(np.mean(stat))}")
    print(f"Median: {np.median(stat)}")
    print(f"Standard Deviation: {round(np.std(stat))}")
    print(f"Min: {np.min(stat)}")
    print(f"Max: {np.max(stat)}")

print("---Basic Descriptive Statistics on age---")
descipt_stat(age) # must put a print statement before it
#statement = print("---Basic Descriptive Statistics on ---")
print("---Basic Descriptive Statistics on sex---")
descipt_stat(sex)
print("---Basic Descriptive Statistics on trestpbs")
descipt_stat(trestbps)
print("---Basic Descriptive Statistics on chol---")
descipt_stat(chol)
print("---Basic Descriptive Statistics on fbs---")
descipt_stat(fbs)
print("---Basic Descriptive Statistics on restecg---")
descipt_stat(restecg)
print("---Basic Descriptive Statistics on thalach---")
descipt_stat(thalach)
print("---Basic Descriptive Statistics on exang---")
descipt_stat(exang)
print("---Basic Descriptive Statistics on oldpeak---")
descipt_stat(oldpeak)
print("---Basic Descriptive Statistics on slope---")
descipt_stat(slope)
print("---Basic Descriptive Statistics on ca---")
descipt_stat(ca)
print("---Basic Descriptive Statistics on thal---")
descipt_stat(thal)
print("---Basic Descriptive Statistics on target---")
descipt_stat(target)

print(f"Number of rows before filtering: {len(data)}")
```

```python
#filtering 

print(f"Number of rows before filtering: {len(data)}")
filtered_data0=data[(target==1)]
print(f"Number of rows after filtering for people who have heart disease {len(filtered_data0)} ")
filtered_data4=data[(age>=40)]
print(f"Number of rows after filtering for people 40 or older {len(filtered_data4)} ")
filtered_data1=data[(age>=40)&(target==1)]
print(f"Number of rows after filtering for people 40 or older and have heart disease: {len(filtered_data1)}")
filtered_data2=data[(age>=40)&(target==1)&(cp>=2)]
print(f"Number of rows after filtering for people 40 or older and have heart disease and have a max chest pain rating of 2 or greater: {len(filtered_data2)}")
filtered_data3=data[(chol>=240) & (target==1)]
print(f"Number of rows after filtering for people who have over 240 cholesterol level and heart disease: {len(filtered_data3)}")
filtered_data5=data[(chol>=240) & (target==1)&(age>=40)]
print(f"Number of rows after filtering for people who have over 240 cholesterol level and heart disease and are 40 years of age or older: {len(filtered_data5)}")
print(f"according to the sample {(round(float(len(filtered_data2))/len(filtered_data1)*100))}% of people who are over the age of 40 and have heart disease have a chest pain rating of a 2 or above.")



#unique categories
print(f"Unique heart disease counts: {len(np.unique(target))}")
print(f"Unique age counts: {len(np.unique(age))}")
print(f"Unique cholesterol counts: {len(np.unique(chol))}")
print(f"Unique chest pain counts: {len(np.unique(cp))}")

#correlations

#target and age
#target and chol
#target and cp 

corr_age=np.corrcoef(target,age,rowvar=False)
#print(corr_age)
print(f"Correlation between heart disease and age is -22% (weak negative).")
corr_chol=np.corrcoef(target,chol>=240,rowvar=False)
#print(corr_chol)
print("Correlation between heart disease and high cholesterol is -11% (weak negative).")
corr_cp=np.corrcoef(target,cp,rowvar=False)
#print(corr_cp)
print("Correlation between heart disease and high cholestorol is 43% (moderate positive).")

print("Based on these results we have found that there is a weak negative correlation between having heart disease and your age or cholesterol level.")
print("What seems to be the most common and reliable correlation to heart disease is chest pain as it has a 43% correlation")
print("Based on this statistical analysis I would reccomend that if you are experiencing chest pain then I would go to the doctors to check for potentinal heart disease.")
```
